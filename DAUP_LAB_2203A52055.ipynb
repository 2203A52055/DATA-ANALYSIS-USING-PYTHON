```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns # Added import for seaborn

file_path = "/content/sample_data/california_housing_test.csv"
df = pd.read_csv("/content/sample_data/california_housing_test.csv")

# Plot scatter plots for all numerical features
sns.pairplot(df, diag_kind="kde")
plt.show()
```

```python
plt.figure(figsize=(8,7))
```

```python
plt.figure(figsize = (8, 6))
plt.scatter(df['households'], df['median_house_value'])
plt.xlabel('households')
plt.ylabel('median_house_value')
plt.title('Scatter Plot of households vs. Median House Value')
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['population'])
plt.xlabel('Population')
plt.title('Box Plot of population')
plt.show()
```
